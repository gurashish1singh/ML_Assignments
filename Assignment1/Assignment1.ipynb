```python
# Part 4

# Q1 - cerating a series and replacing characters

series3 = pd.Series(['A','B','C','AabX','BacX','',np.nan,'CABA','dog','cat'])
print('Original Series: \n', series3)
# series3.replace(to_replace = ['dog'], value = ['XX-XXX']) # Replacing dog
series3.str.replace('x|dog','XX-XXX', case = False)  # Replacing X
# series3.replace('X|dog','XX-XXX')  # Replacing X
```

```python
# Q2 - removing characters from series

series4 = pd.Series(['12','-$10','$10,000'])
series4
series4.str.replace('$','')
```

```python
# Q3 - reversing all lower case

series5 = pd.Series(['france 1998', 'country', np.nan])
print'Original Series \n', series5
repl = lambda m: m.group(0)[::-1]
series5.str.replace(r'[a-z]+', repl)
```

```python
df = pd.read_csv('SFO Public Department.csv')
df.describe()
```

```python
df.sort_values('Year')
```

```python
# 1 - calclating the salary increase
pay_2015=df.loc[df['Year'] == 2015]
print(pay_2015.TotalPayBenefits.sum())

pay_2016=df.loc[df['Year'] == 2016]
print(pay_2016.TotalPayBenefits.sum())

pay_2017=df.loc[df['Year'] == 2017]
print(pay_2017.TotalPayBenefits.sum())

pay_2018=df.loc[df['Year'] == 2018]
print(pay_2018.TotalPayBenefits.sum())

salary_increase = pay_2018.TotalPayBenefits.sum() - pay_2015.TotalPayBenefits.sum()
print'Salary Increase = ' ,salary_increase
```

```python
# 2 - highest mean salary for job title in 2014
pay_2018.groupby('JobTitle').mean()['TotalPayBenefits'].sort_values(ascending=False).head(1)
# pay_2018.JobTitle[pay_2018['TotalPayBenefits'] == pay_2018['TotalPayBenefits'].apply('mean')]
```

```python
# 3 - money saved by stopping OverTimePay
print pay_2018.TotalPayBenefits.sum()
print pay_2018.OvertimePay.sum()
print 'Money Saved = ', pay_2018.TotalPayBenefits.sum()-pay_2018.OvertimePay.sum()
```
